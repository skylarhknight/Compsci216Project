<h1>Exploratory Data Analysis</h1>


```python
#fatalities per factor
#basically just made the colors lighter/darker based on amount of fatalities

# converting any commas in the DataFrame to dots and convert to numeric
crash_data = crash_data.replace(',', '', regex=True).astype(int)

# color and brightness
def modify_brightness(color, brightness_factor):
    hsv = plt.cm.colors.rgb_to_hsv(color[:3])
    hsv[2] = hsv[2] * brightness_factor
    rgb = plt.cm.colors.hsv_to_rgb(hsv)
    return rgb


original_cmap = plt.get_cmap('Blues')
brightness_factor = 0.9  

# brightness of each color in the original colormap
modified_colors = [modify_brightness(original_cmap(i), brightness_factor) for i in range(original_cmap.N)]

# new colormap
modified_cmap = plt.cm.colors.ListedColormap(modified_colors)

# normalizing the 'Fatal(K)' values to range from 0 to 1
normalize = plt.Normalize(vmin=crash_data['Fatal(K)'].min(), vmax=crash_data['Fatal(K)'].max())

# plotting
plt.figure(figsize=(12, 8))
bars = plt.bar(crash_data.index, crash_data['Fatal(K)'], color=modified_cmap(normalize(crash_data['Fatal(K)'])))
plt.title('Fatal Injuries for Each Variable')
plt.xlabel('')
plt.ylabel('Fatal Injuries')
plt.xticks(rotation=90)
plt.grid(True)


sm = plt.cm.ScalarMappable(cmap=modified_cmap, norm=normalize)
sm.set_array([])
cbar = plt.colorbar(sm)
cbar.set_label('Normalized Fatal(K)')

plt.show()
```

```python
#using rows as independent data
crash_data_transposed = crash_data.transpose()

#correlation matrix
correlation_matrix = crash_data_transposed.corr()

# heatmap
# kind of hard to read but still has good info
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap')
plt.show()

```

```python
#correlation amongst the independent variables 


crash_data_Filtered = crash_data.drop(columns=['Unknown'])

corr_matrix = crash_data_Filtered.corr()
plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title('Correlation Matrix')
plt.show()
```

```python
#scatterplot 

plt.figure(figsize=(8, 6))
sns.scatterplot(data=crash_data, x='No Injury(O)', y='Fatal(K)')
plt.title('No Injuries vs. Fatal Injuries')
plt.grid(True)
plt.show()
```

```python
#analyzing age correlation
subset_data = crash_data.iloc[0:6]

# Create a scatter plot using the index as one of the variables
plt.figure(figsize=(8, 6))
plt.scatter(subset_data.index, subset_data['Fatal(K)'])
plt.title('Index vs. Fatal Injuries')
plt.xlabel('Index')
plt.ylabel('Fatal Injuries')
plt.grid(True)

# Rotate the x-axis tick labels by 45 degrees
plt.xticks(rotation=90)

plt.show()
```

<h1>Hypothesis Testing</h1>

```python
data = crash_data
data = pd.DataFrame({
    'Demographic Factor': ['MALE', 'Alcohol Not Involved', 'PASSENGER CAR'],  # Replace with your actual categories
    'Fatal': [2703, 3438, 1068]  # Replace with actual counts
})

# contingency table using crosstab
contingency_table = pd.crosstab(index=data['Demographic Factor'], 
                                columns='Fatal Count', 
                                values=data['Fatal'], 
                                aggfunc='sum')


print(contingency_table)
```

```python
# chi-square test of independence
chi2, p_value, dof, expected = chi2_contingency(contingency_table)

print("Chi-square statistic:", chi2)
print("P-value:", p_value)
```
